```python
from travelAndMutate.randominterface import NumpyRandomGenerator
#from travelAndMutate.types import *
from travelAndMutate.system import System
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from time import time
```

```python
rng = NumpyRandomGenerator()
```

```python
Npatch = 2
```

```python
#%%timeit

start = time()

commuting = np.where(np.eye(Npatch).astype(bool), 1, rng.rng.random(size=(Npatch,Npatch))/Npatch**2)
sys = System(rng.cpprng, commuting, 1)

Ns = np.full(Npatch, 1e3, dtype=int)
betas = np.full(Npatch, 1)# + rng.rng.normal(scale=0.01, size=Npatch)
epsilons = np.full(Npatch, 1)# + rng.rng.normal(scale=0.005, size=Npatch)
mus = np.full(Npatch, 0.5)# + rng.rng.normal(scale=0.005, size=Npatch)
for i in range(Npatch):
    sys.setPatchProperties(i, Ns[i], betas[i], epsilons[i], mus[i])
print(f"<R0> = {np.mean(betas/mus*commuting.sum(axis=1))}")

sys.seedEpidemic(np.full(Npatch,100,dtype=int).tolist())

sys.spreadForTime(10)

end = time()

print(f"Elapsed: {end-start} sec")
#print(".", end="", flush=True)
```
